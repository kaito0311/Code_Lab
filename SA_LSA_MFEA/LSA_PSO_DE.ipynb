```python
from utils.benchmark_function import *
from SaMTPSO.lsa_SaMTPSO_DE import * 

tasks = getManyTasks10()

history_cost, history_p_matrix = [], [] 
for i in range(1):

    history, p_matrix = lsa_SaMTPSO_DE(tasks, lsa= True)
    history_cost.append(history)
    history_p_matrix.append(p_matrix) 
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize= (20,10)) 
p_matrix = history_p_matrix[0]
for i in range(len(tasks)):
    plt.plot(np.arange(len(p_matrix)), [matrix[6-1][i] for matrix in p_matrix], label = "task " + str(i+1))
    plt.legend()
```

```python
name = ["sa", "lsa", "mfea"]
fig = plt.figure(figsize = (50, 30), dpi= 100)
fig.set_facecolor("white")
color = ['r', 'b', 'y']
history = [history_cost[0]]
for task in range(NUMBER_TASKS):
    for version in range(len(history)):
        x_axis = np.arange(len(history[version]))
        y_axis =  [x[task].cost for x in history[version]]

        plt.subplot(2, int(len(tasks)/2), task+1)
        plt.plot(x_axis, y_axis, color[version], label = name[version])
        plt.yscale('log')
        plt.legend()

    plt.title(tasks[task].name)
plt.show()
```
